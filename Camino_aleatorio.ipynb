Caminos aleatorios.

En el mundo aleatorio, es posible definir un camino. Esto es posible, si se construye un variable que almacene los valores aleatorios en un arreglo,  ademas suponiendo un espacio bidimensional, que esté formado con base (x,y), donde x+dx, y +dy esté construido de manera aleatoria, entonces el camino resultante es de naturaleza aleatoria

```python
from random import random, seed
import matplotlib.pyplot as plt
from IPython import display
import time
import numpy
```

```python
N = 100 # cantidad de datos aleatorios
seed(5) # semilla
# componentes del camino aleatorio
x = 0 # DOMINIO
y = 0 # RANGO

R = numpy.zeros((N,2)) # vector de posiciones
plt.figure(figsize=(8,5))
for i in range(N):
    x += 2*(random() - 0.5) # aumenta el paso x + dx
    y += 2*(random() - 0.5) # aumenta el paso y + dy
    R[i,0] = x # Guarda el aumento en el dominio
    R[i,1] = y # Guarda al aumento den el rango
    # grafica de x vs y
    plt.gca().cla()
    plt.plot(R[0:i,0],R[0:i,1]) 
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(0.001)
```

Dado que la semilla proporciona el caracter aleatorio, es posible contruir diferentes caminos aleatorios para diferentes semillas

```python
# funcion camino
def walk(N, semilla):
    
    seed(semilla) # semilla
    x = 0
    y = 0 # componentes aleatorias inicail
    R = numpy.zeros((N,2)) # vector de posiciones
    for i in range(N):
        x += 2*(random() - 0.5)
        y += 2*(random() - 0.5)
        R[i,0] = x
        R[i,1] = y
    return (R)
```

```python
N = 10000 # cantidad de datos aleatorios
plt.figure(figsize=(8,5))
for semilla in range(3): # cantidad de caminos, con semilla diferente
    R = walk(N,semilla)
    plt.plot(R[:,0],R[:,1])
```
